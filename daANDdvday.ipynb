## Search what is histogram

```python
#histogram
data1['SalePrice'].hist(bins=30)
# try to change the bin numbers 
```

```python
#skewness and kurtosis? WHat are they?
#skewness-is the lack of symmetry
#kurtosis-to understand the data are heavy-tailed or light-tailed relative to a normal distribution
print("Skewness: %f" % data1['SalePrice'].skew())
print("Kurtosis: %f" % data1['SalePrice'].kurt())
```

```python
#scatter plot grlivarea/saleprice
var = 'GrLivArea'
data1 = pd.concat([data1['SalePrice'], data1[var]], axis=1)
data1.plot.scatter(x=var, y='SalePrice', ylim=(0,500000));
```

```python
#import library for making statistical graphics
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
f = plt.figure(figsize=(19, 15))
plt.matshow(data1.corr())
plt.show()
```

```python
#correlation matrix
#here we can is the value higher
#if value is higher then the color is brighter
corrmat = data1.corr()
f, ax = plt.subplots(figsize=(12, 9))
sns.heatmap(corrmat, vmax=.8, square=True);
```
